### Approach

The goal is to identify the best predictive model for loan approval and use it to estimate an applicant's approval probability, deriving a proxy risk score for credit risk segmentation. 

**Note:** the target variable is loan approval status, not observed default — true PD estimation requires post-approval repayment data which is not available in this dataset.

The following steps are performed:

- Building a Logistic regression model with and without hyperparameter tuning on cleaned dataset
- Building an XGBoost model on the original unprocessed dataset to demonstrate its native ability to handle missing values and skewed features without preprocessing — a deliberate methodological choice to contrast with the preprocessed logistic regression data
- Comparing models using accuracy, ROC-AUC score, f1 score, and ROC-AUC curves and deciding the best model for these metrics.
- Evaluate credit risk using best model and categorizing applicant in to three groups Low Risk - Medium Risk - High Risk based Probability of default.
- Logistic Regression selected as final model despite comparable performance. IRB frameworks require directly interpretable coefficients for regulatory compliance and auditability

### Business Context

Financial institutions use credit risk modeling to assess the likelihood of borrowers repaying loans. Poor lending decisions increase default risk and financial losses. By analyzing historical loan application data, predictive models identify patterns in borrower characteristics and financial indicators that influence loan approval.

In this project, the objective is to explore borrower characteristics and financial indicators that influence loan approval decisions and to develop a model that estimates the likelihood of loan approval.

## 4.1 Logistic Regression

Compared models trained with default hyperparameters to models with optimized hyperparameters. This demonstrats how tuning improves predictive performance, model robustness, and generalization, providing evidence for selecting the most effective model for the data.

I will be dropping the original income and loan data for model building, so they wouldn't affect the model from containing outliers etc. These featuers will be needed later of calculating expected loss. Data will also be scaled using z-score standardization.

```python
X_train = X_train_log.copy().drop(["Applicant_Income", "Coapplicant_Income", "Loan_Amount", 
                                   "Total_Income", "DTI", "Dept_Ratio"], axis=1)
y_train = y_train_log.copy().squeeze()
X_test = X_test_log.copy().drop(["Applicant_Income", "Coapplicant_Income", "Loan_Amount", 
                                   "Total_Income", "DTI", "Dept_Ratio"], axis=1)
y_test = y_test_log.copy().squeeze()
```

```python
numeric_cols = X_train.select_dtypes(include="float64").columns

scaler = StandardScaler()

X_train[numeric_cols] = scaler.fit_transform(X_train[numeric_cols])
X_test[numeric_cols] = scaler.transform(X_test[numeric_cols])
```

### 4.1.1 Without hyperparametertunning

```python
logreg = LogisticRegression(max_iter=1000,random_state = 42)
logreg.fit(X_train, y_train)

y_pred = logreg.predict(X_test)
y_pred_prob = logreg.predict_proba(X_test)[:,1]
auc_logreg = roc_auc_score(y_test, y_pred)
accuracy = accuracy_score(y_test, y_pred)
f1score = f1_score(y_test, y_pred)

print("Accuracy:", accuracy)
print("ROC AUC Score:", auc_logreg)
print("F1-score:", f1score, "\n")
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred),"\n")
print("Classification Report:\n", classification_report(y_test, y_pred))
```

### 4.1.2 With hyperparametertunning

Logistic regression hyperparameters were optimized using GridSearch with 5-fold cross-validation.

The regularization parameter C was tested across logarithmic values to balance bias and variance.
Both L1 and L2 penalties were evaluated to compare sparse vs stable coefficient estimation.

The ROC AUC metric was used because credit datasets are typically imbalanced, and AUC better measures the model’s ability to distinguish defaulters from non-defaulters.

```python
# Define parameter grid
param_grid = [
    {
        "penalty": ["l1"],
        "solver": ["liblinear","saga"],
        "C": [0.01,0.1,1,10,100]
    },
    {
        "penalty": ["l2"],
        "solver": ["lbfgs","liblinear","saga"],
        "C": [0.01,0.1,1,10,100]
    }
]

grid_search = GridSearchCV(
    LogisticRegression(random_state=42, max_iter=1000),
    param_grid[0],
    scoring="roc_auc",
    cv=5,
    n_jobs=-1
)

# Fit on training data
grid_search.fit(X_train, y_train)

print("Best Parameters:", grid_search.best_params_, "\n")
print("Best ROC AUC (CV):", grid_search.best_score_)
```

# 5. Risk Segmentation

Due to limitations in the dataset, including the absence of observed repayment outcomes, traditional credit risk parameters such as Probability of Default (PD), Exposure at Default (EAD), and Loss Given Default (LGD) could not be fully estimated. In particular, no information on borrower repayment behaviour is available, making true PD modelling infeasible.

As a result, the model developed in this project predicts loan approval probability rather than default risk. To approximate credit risk, a risk score was derived as (1 − Probability of Approval). This score is interpreted as a proxy for borrower risk rather than a true PD. In a real-world setting, a separate PD model would be developed using only approved loans with observed repayment outcomes.

Risk-related features such as Debt-to-Income (DTI) and Loan-to-Income (LTI) ratios were incorporated to capture repayment capacity and financial stress. For simplicity, all exposures were assumed to be senior unsecured, and a conservative Loss Given Default (LGD) of 50% was applied as a proxy assumption (Basel II Glossary, 2013: https://web.archive.org/web/20130421005422/http://www.basel-ii-risk.com/Basel-II/Basel-II-Glossary/Loss-Given-Default.htm).

Based on the derived risk score, applicants were segmented into three risk categories:

- **Low Risk (Score < 0.10):** Applicants with high likelihood of approval and low estimated risk

- **Medium Risk (0.10 ≤ Score < 0.50):** Borderline applicants recommended for manual review

- **High Risk (Score ≥ 0.50):** Applicants with high estimated risk, typically rejected

This segmentation is supported by model performance observed in the confusion matrix:

- **True Negatives (TN = 84):** Eligible applicants correctly approved, corresponding primarily to the Low Risk segment

- **False Negatives (FN = 17):** Ineligible applicants incorrectly approved, concentrated in the Medium Risk segment, highlighting the importance of manual review

- **True Positives (TP = 21):** Ineligible applicants correctly rejected, aligning with the High Risk segment

- **False Positives (FP = 1):** Eligible applicants incorrectly rejected, indicating the model remains conservative but not overly restrictive

**Interpretation:**
The model applies a conservative decision framework, prioritising minimisation of high-risk approvals while maintaining a high approval rate for low-risk applicants. The use of model-driven risk segmentation enables identification of borderline cases and supports explainable decision-making, even in the absence of true default data.

```python
# 1. Predict probability of approval
probabilities = logreg_h.predict_proba(X_test)[:, 1]
X_test_r["Probability_Approved"] = probabilities

# 2. Actual outcome
X_test_r["Loan_Approved"] = y_test_log.values

# 3. Risk score (NOT PD)
X_test_r["Risk_Score"] = 1 - X_test_r["Probability_Approved"]

# 4. Add rejected column (important for validation)
X_test_r["Rejected"] = 1 - X_test_r["Loan_Approved"]

# 5. Risk categories
def risk_bucket(score):
    if score <= 0.10:
        return "Low Risk"
    elif score <= 0.50:
        return "Medium Risk"
    else:
        return "High Risk"

X_test_r["Risk_Category"] = X_test_r["Risk_Score"].apply(risk_bucket)

# 6. Expected Loss (proxy)
LGD = 0.5
X_test_r["Expected_Loss"] = X_test_r["Risk_Score"] * X_test_r["Loan_Amount"] * LGD

# 7. Summary table
summary = X_test_r.groupby("Risk_Category").agg(
    Num_Applicants=("Loan_Amount", "count"),
    Avg_Risk_Score=("Risk_Score", "mean"),
    Total_Loan_Amount=("Loan_Amount", "sum"),
    Total_Expected_Loss=("Expected_Loss", "sum")
).reset_index()

summary["Avg_Risk_Score"] = (summary["Avg_Risk_Score"] * 100).round(1)
summary["Total_Expected_Loss"] = summary["Total_Expected_Loss"].round(0)

summary
```
